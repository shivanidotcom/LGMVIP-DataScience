# **Stock Market Prediction And Forecasting Using Stacked LSTM**


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
from sklearn.preprocessing import MinMaxScaler
```

```python
data=pd.read_csv('/content/dataset.csv')
```

```python
data.head()
```

```python
data.describe()
```

```python
#checking for null values
data.isnull()
```

```python
df=data.sort_values(by='Date')
df.head()
```

```python
df.reset_index(inplace=True)
df.head()
```

```python
plt.plot(df['Close'])
dff=df['Close']
dff
```

```python
training_size=int(len(dff)*0.70)
test_size=len(dff)-training_size
train_data,test_data=dff[0:training_size,:],dff[training_size:len(dff),:1]
```

```python
def create_dataset(dataset, time_step=1):
    dataX, dataY = [], []
    for i in range(len(dataset)-time_step-1):
        a = dataset[i:(i+time_step), 0]   
        dataX.append(a)
        dataY.append(dataset[i + time_step, 0])
    return np.array(dataX), np.array(dataY)
```

```python
#spliting the dataset
time_step = 100
X_train, y_train = create_dataset(train_data, time_step)
X_test, ytest = create_dataset(test_data, time_step)
```

```python
X_train =X_train.reshape(X_train.shape[0],X_train.shape[1] , 1)
X_test = X_test.reshape(X_test.shape[0],X_test.shape[1] , 1)
```

```python
#Now creating the stacked LSTM MODEL
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.layers import LSTM
model=Sequential()
model.add(LSTM(50,return_sequences=True,input_shape=(100,1)))
model.add(LSTM(50,return_sequences=True))
model.add(LSTM(50))
model.add(Dense(1))
model.compile(loss='mean_squared_error',optimizer='adam')
model.summary()
```

```python
model.fit(X_train,y_train,validation_split=0.1,epochs=60,batch_size=64,verbose=1)
```

```python
test_predict=model.predict(X_test)
```

```python
import math
from sklearn.metrics import mean_squared_error
```

```python
performance = math.sqrt(mean_squared_error(ytest,test_predict))
performance
```
